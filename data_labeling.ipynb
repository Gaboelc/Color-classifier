### Importing libraries

```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
```

### Loaging the data

```python
df = pd.read_csv('data/wikipedia_color_names.csv')
df
```

### Checking if there are any missing values in the data

```python
df.isnull().sum()
```

### Getting just the RGB values from the data

```python
data = df[['Red (8 bit)', 'Green (8 bit)', 'Blue (8 bit)']].values
data
```

### Importing A color classifier model

```python
color_classifier = pickle.load(open('Models/color_classifier.sav', 'rb'))
```

### Labeling new data

```python
results = color_classifier.predict(data)
df['Color Type'] = results
df
```
